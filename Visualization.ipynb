```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
data = pd.read_csv("/content/FINALENERGYCSV.csv")
```

```python
data.head()
```

```python
data[data["Vector"] == 0]["PMTN"].hist(label="Vector=0")
data[data["Vector"] == 1]["PMTN"].hist(label="Vector=1")
plt.legend()
plt.xlabel("PMT Number")
plt.ylabel("Frequency")
plt.title("Distribution of PMTs for Vectors with 0 and 1")
plt.show()
```

```python
sns.lineplot(x="PMTN", y="Energy", data=data)
plt.xlabel("PMT Number")
plt.ylabel("Energy")
plt.title("Energy Distribution")
plt.show()
```

```python
sns.lineplot(x="PMTN", y="Time", data=data)
plt.xlabel("PMT Number")
plt.ylabel("Time")
plt.title("Time Distribution")
plt.show()
```

```python
fig, ax1 = plt.subplots(figsize=(10, 6))

sns.lineplot(x="PMTN", y="Time", data=data, ax=ax1)

#second axis
ax2 = ax1.twinx()
sns.lineplot(x="PMTN", y="Energy", data=data, ax=ax2, color = "red")

ax1.set_xlabel("PMT Number")
ax1.set_ylabel("Time")
ax2.set_ylabel("Energy", color = "red")
plt.title("PMT Energy and Time Relationship")

plt.tight_layout()
plt.show()
```
