```python
import pandas as pd
import numpy as np
```

```python
data = pd.read_csv("/content/data.csv")
```

```python
vector_1 = data[data['Vector'] == 1]
vector_1
```

```python
vector_0 = data[data['Vector'] == 0]
vector_0
```

```python
balanced_data = pd.concat([vector_0.sample(35000), vector_1.sample(38000)], axis=0)
balanced_data
```

```python
balanced_data = balanced_data.sample(frac=1).reset_index(drop=True)
balanced_data
```

```python
balanced_data.to_csv("balanced_data.csv")
```
